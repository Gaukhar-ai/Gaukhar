```
#start with loading all necessary libraries
import numpy as np
#numpy provides fast math function processing
import pandas as pd
#pandas is used for easy indexing and data collection tools
from PIL import Image
#pillow is a very good Py image library
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
#Im importing necessary packages for word cloud
import matplotlib.pyplot as plt
% matplotlib inline
```

```
#second, i manually add words from my resume
text=("Diamond Gaukhar Javarova Diamond Gaukhar Javarova Machine Learning Algorythms Stock Market Trading Gaukhar Javarova Diamond Python Algos Statistics Mathematics Data Science ")
```

```
#third I create the wordcloud object
wordcloud = WordCloud(width=850, height=450, margin=10).generate(text)
```

```
#Fourth I insert the code to display the generated image:
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.margins(x=30, y=0)
plt.show()
```
